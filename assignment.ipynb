```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from urllib.parse import urljoin

# ============================================
# PART A: WEB SCRAPING AND DATA COLLECTION
# ============================================

def scrape_books(base_url="https://books.toscrape.com/"):

    all_books = []

    rating_map = {
        "One": 1,
        "Two": 2,
        "Three": 3,
        "Four": 4,
        "Five": 5
    }

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    # Reuse HTTP connection (faster)
    session = requests.Session()

    print("=" * 60)
    print("Starting Web Scraping...")
    print("=" * 60)

    for page in range(1, 51):

        url = f"{base_url}catalogue/page-{page}.html"

        try:

            response = session.get(url, headers=headers, timeout=10)

            if response.status_code != 200:
                print(f"Page {page} not found.")
                break

            soup = BeautifulSoup(response.text, "html.parser")

            books = soup.find_all("article", class_="product_pod")

            if not books:
                break

            for book in books:

                # Title
                title = book.h3.a["title"]

                # Price
                price_text = book.find("p", class_="price_color").text
                price = float(re.sub(r"[^\d.]", "", price_text))

                # Rating
                rating_text = book.find("p", class_="star-rating")["class"][1]
                rating_numeric = rating_map.get(rating_text, 0)

                # Availability
                availability = book.find(
                    "p",
                    class_="instock availability"
                ).get_text(strip=True)

                # Product Link
                relative_link = book.h3.a["href"]
                product_link = urljoin(
                    base_url + "catalogue/",
                    relative_link
                )

                all_books.append({
                    "title": title,
                    "price_gbp": price,
                    "rating_numeric": rating_numeric,
                    "availability": availability,
                    "product_link": product_link,
                    "page": page
                })

            print(f"✓ Page {page} completed ({len(all_books)} books)")

        except Exception as e:
            print(f"Error on page {page}: {e}")
            break

    print("\nScraping completed successfully!")
    print(f"Total Books Scraped: {len(all_books)}")

    return pd.DataFrame(all_books)


# ============================================
# Execute Scraping
# ============================================

df_scraped = scrape_books()

# ============================================
# Display Dataset
# ============================================

print("\nFirst 5 Records")
display(df_scraped.head())

print("\nDataset Shape")
print(df_scraped.shape)

print("\nDataset Information")
print(df_scraped.info())

print("\nSummary Statistics")
display(df_scraped.describe())

# ============================================
# Save Raw Dataset
# ============================================

df_scraped.to_csv(
    "scraped_raw_dataset.csv",
    index=False,
    encoding="utf-8"
)

print("\n✓ Dataset saved as 'scraped_raw_dataset.csv'")

# ============================================
# Basic Analysis
# ============================================

print("\nTotal Books :", len(df_scraped))
print("Unique Books :", df_scraped["title"].nunique())
print("Average Price : £", round(df_scraped["price_gbp"].mean(), 2))
print("Maximum Price : £", df_scraped["price_gbp"].max())
print("Minimum Price : £", df_scraped["price_gbp"].min())

print("\nRating Distribution")
print(df_scraped["rating_numeric"].value_counts().sort_index())

print("\nAvailability")
print(df_scraped["availability"].value_counts())
```
